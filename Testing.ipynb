```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Load and display the saved figure
plt.figure(figsize=(12, 6))
img = plt.imread('motion_detection_output/motion_detection_result.png')
plt.imshow(img)
plt.axis('off')
plt.show()

# Load and print some information about the predictions
predictions = np.load('motion_detection_output/predictions.npy')
print(f"Number of lines in the image: {len(predictions)}")
print(f"Number of lines detected as motion-affected: {np.sum(predictions > 0.5)}")
print(f"Average prediction value: {np.mean(predictions):.4f}")

# Optionally, plot the prediction values
plt.figure(figsize=(10, 4))
plt.plot(predictions)
plt.title('Prediction Values for Each Line')
plt.xlabel('Line Index')
plt.ylabel('Prediction Value')
plt.ylim(0, 1)
plt.show()
```
